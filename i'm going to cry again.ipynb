```python
list2 = doc.text.split('"')
```

```python
indices = [i for i, x in enumerate(list2) if x == "track"]
```

```python
indices
```

```python
x=0
rows = []

while x<len(indices):
    row = {}
    row['artist_name'] = (list2[indices[x]+32]) #32 is artist
    row['song_name'] = (list2[indices[x]+62]) # 62 is song name
    row['playcount'] = int(list2[indices[x]+74]) #74 is playcount
    print(row)
    rows.append(row)
    x=x+1

dataTime = pd.DataFrame(rows)
```

```python
dataTime['percent'] = (dataTime['playcount']/(sum(dataTime['playcount'])))*100
dataTime['percent'] = round(dataTime['percent'], 0)
```

```python
rows2 = []
song_names = dataTime['song_name']
percentages = dataTime['percent']
rows2.append(song_names)
rows2.append(percentages)
dataTime2 = pd.DataFrame(rows2)
dataTime2
```

```python
df3 = dataTime[['song_name', 'percent']]
df3 = df3.sort_values(by=['percent'], ascending = False)
```

```python
len(df3)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plotting the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Start position for each segment in the stacked bar
start_position = 0

# Assign a unique color for each segment
colors = plt.cm.Paired(range(len(df3)))

# Loop through each row in the dataframe and add to the stacked bar
for index, row in df3.iterrows():
    ax.barh('Total', row['percent'], left=start_position, color=colors[index], label=row['song_name'])
    start_position += row['percent']

# Add labels and title
ax.set_xlabel('Percent')
ax.set_title('Percentages of Total Plays')

# Add song labels to the legend
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
